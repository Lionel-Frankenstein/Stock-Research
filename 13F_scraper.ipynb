```python
from selenium import webdriver
from tqdm import tqdm

tickers = []
names = []
ratios = []
buys = []
hold_prices = []
current_prices = []
year_lows = []
above_year_lows = []
year_highs = []

driver = webdriver.Chrome()

for i in range(1, 11):
    url = f"https://www.dataroma.com/m/g/portfolio_b.php?q=q&o=c&L={i}"
    driver.get(url)
    trs = driver.find_elements(webdriver.common.by.By.TAG_NAME, "tr")
    for tr in tqdm(trs[1:-1]):
        tds = tr.find_elements(webdriver.common.by.By.TAG_NAME, "td")
        tickers.append(tds[0].text)
        names.append(tds[1].text)
        ratios.append(tds[2].text)
        buys.append(tds[3].text)
        hold_prices.append(tds[4].text)
        current_prices.append(tds[5].text)
        year_lows.append(tds[6].text)
        above_year_lows.append(tds[7].text)
        year_highs.append(tds[8].text)
    
driver.quit()
```

```python
import yahoo_fin.stock_info as si
import pandas as pd
```

```python
quotes_dic = {}
quotes_dic2 = {}
for ticker in tickers[:1]:
    print(ticker)
    stats = si.get_stats(ticker)
    stats2 = si.get_stats_valuation(ticker)
    quotes_dic[ticker] = stats
    quotes_dic2[ticker] = stats2
```

```python
alist = list(stats['Attribute'])
alist2 = list(stats2[0])
columns = ["Ticker", "Name", "% of Portfolio", 'Buys', 'Hold Price', 'Current Price', '52 Week Low', '% Above 52 Week Low', '52 Week High'] + alist + alist2
columns
```

```python
df = pd.DataFrame({col: [] for col in columns})
df
```

```python
error_list = []

for i in tqdm(range(len(tickers))):
    ticker = tickers[i]
    try:
        stats = si.get_stats(ticker)
        stats2 = si.get_stats_valuation(ticker)
        row = [ticker, names[i], ratios[i], buys[i], hold_prices[i], current_prices[i], year_lows[i], above_year_lows[i], year_highs[i]] + list(stats['Value']) + list(stats2[1])
        df.loc[len(df)] = row
    except:
        error_list.append(ticker)
        continue
```
