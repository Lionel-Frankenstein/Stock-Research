```python
from selenium import webdriver

fund_codes = []

# Start a Chrome driver
driver = webdriver.Chrome()

# Navigate to a URL
driver.get('https://www.dataroma.com/m/home.php')

# Find a span element with a specific id
span = driver.find_element("id", "port_body")

# Find all the links under the span element
links = span.find_elements(webdriver.common.by.By.TAG_NAME, "a")

for link in links:
    print(link.get_attribute('href'))
    fund_code = link.get_attribute('href')[42:]
    fund_codes.append(fund_code)

driver.quit()
```

```python
fund_codes
```

```python
stocks_dic = {}
```

```python
for code in fund_codes:
    print(code)
    url = "https://www.dataroma.com/m/holdings.php?m=" + code
    driver = webdriver.Chrome()
    driver.get(url)
    table = driver.find_element("id", "grid")
    trs = table.find_elements(webdriver.common.by.By.TAG_NAME, "tr")
    for tr in trs:
        tds = tr.find_elements(webdriver.common.by.By.TAG_NAME, "td")
        activity = tds[3].text
        if "Buy" in activity or "Add" in activity:
            quote = tds[1].find_element(webdriver.common.by.By.TAG_NAME, "a").text
            if quote in stocks_dic:
                stocks_dic[quote] += 1
            else:
                stocks_dic[quote] = 1
    driver.quit()
```

```python
stocks_dic = dict(sorted(stocks_dic.items(), key=lambda item: item[1], reverse = True))
stocks_dic
```

```python
funds_bought = list(stocks_dic.values())
funds_bought
```

```python
quotes = []
for stock in stocks_dic:
    idx = stock.index(" ")
    quote = stock[:idx]
    quotes.append(quote)
quotes
```
